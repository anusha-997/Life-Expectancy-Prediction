The dataset related to life expectancy, health factors for 183 countries has been collected from the same WHO data repository website and its corresponding economic data was collected from United Nation website. # **Missing Value Handling**

*On initial visual inspection of the data showed some missing values. As the datasets were from WHO, we found no evident errors. The result indicated that most of the missing data was for population, Hepatitis B and GDP. The missing data were from less known countries like Vanuatu, Tonga, Togo,Cabo Verde etc.So by missing value handling technique we have resolved the same. The final dataset consists of 18 Columns and 2928 rows.*

```python
df.isna().sum()
```

```python
df.head()
```

```python
df['Population']=df['Population'].fillna(df['Population'].mode()[0])
df['Hepatitis B']=df['Hepatitis B'].fillna(df['Hepatitis B'].mode()[0])
df['Polio']=df['Polio'].fillna(df['Polio'].mode()[0])
df['Diphtheria']=df['Diphtheria'].fillna(df['Diphtheria'].mode()[0])
df['Total expenditure']=df['Total expenditure'].fillna(df['Total expenditure'].mean())
df['GDP']=df['GDP'].fillna(df['GDP'].median())
df['BMI']=df['BMI'].fillna(df['BMI'].mean())
df['thinness  1-19 years']=df['thinness  1-19 years'].fillna(df['thinness  1-19 years'].median())
df['Alcohol']=df['Alcohol'].fillna(df['Alcohol'].mean())
df['Schooling']=df['Schooling'].fillna(df['Schooling'].mean())
```

# **Separate Features and Targets**

```python
X=df.iloc[:,:-1]
X
```

```python
y=df.iloc[:,-1]
y
```

# **Scaling**

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
X_scaled=scaler.fit_transform(X)
X_scaled
```

# **Split train and test datasets**

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X_scaled,y,test_size=0.3,random_state=1)
```

```python
X_train.shape,y_train.shape
```

```python
X_test.shape,y_test.shape
```

# **Model Creation and Checking Performance measures**

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor,AdaBoostRegressor
from xgboost import XGBRegressor
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
knr=KNeighborsRegressor(n_neighbors=3)
mlr=LinearRegression()
svr=SVR(kernel='linear')
dt=DecisionTreeRegressor(random_state=42)
rf=RandomForestRegressor(random_state=42)
ab=AdaBoostRegressor(random_state=42)
xgb=XGBRegressor()
models=[knr,mlr,svr,dt,rf,ab,xgb]
for model in models:
  print("**************",model,"***************")
  model.fit(X_train,y_train)
  y_pred=model.predict(X_test)
  print("R2 score:",r2_score(y_test,y_pred)*100)
  print("MAE:",mean_absolute_error(y_test,y_pred))
  print("MSE:",mean_squared_error(y_test,y_pred))
  print("RMSE:",np.sqrt(mean_squared_error(y_test,y_pred)))
```

```python
model_name=[]
r2=[]
mae=[]
mse=[]
rmse=[]
# models=[KNeighborsRegressor(),LinearRegression(),SVR(),DecisionTreeRegressor(),RandomForestRegressor(random_state=42),AdaBoostRegressor(),XGBRegressor()]
models=[knr,mlr,svr,dt,rf,ab,xgb]
for model in models :
    model.fit(X_train,y_train)
    y_pred=model.predict(X_test)
    model_name.append(model.__class__.__name__)
    r2.append(str(r2_score(y_test, y_pred)*100))
    mae.append(str((mean_absolute_error(y_test, y_pred))))
    mse.append(str((mean_squared_error(y_test, y_pred))))
    rmse.append(str((mean_squared_error(y_test, y_pred, squared=False))))
models_df=pd.DataFrame({"Model-Name":model_name,"R2":r2,"MAE":mae,"MSE":mse,"RMSE": rmse,})
models_df=models_df.set_index('Model-Name')
models_df.sort_values("R2")
```
